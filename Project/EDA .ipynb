```python
# 1922-1950 144     (I have divided the year into groups for better and easy understandability)
# 1951-2000 607     (counted with the help of MS Excel)
# 2001-2022 913
```

```python
Year_1 = Rate[ ( (Rate['Year'] >= 1922) & (Rate['Year'] <= 1950))]
Year_1
```

```python
Year_2 = Rate[ ( (Rate['Year'] >= 1951) & (Rate['Year'] <= 2000))]
Year_2
```

```python
Year_3 = Rate[ ( (Rate['Year'] >= 2001) & (Rate['Year'] <= 2022))]
Year_3
```

```python
Year1 = Year_1['Rating']
Year2 = Year_2['Rating']
Year3 = Year_3['Rating']
```

```python
y1 =Year1.mean().round(2)
y1
```

```python
y2 =Year2.mean().round(2)
y2
```

```python
y3 =Year3.mean().round(2)
y3
```

```python
Years= ['1922-1950','1951-2000','2001-2022']
Year_rate_mean= [y1,y2,y3]
```

```python
plt.bar(Years,Year_rate_mean )
plt.xlabel('Year')
plt.ylabel('Ratings')
plt.grid()
plt.title('Ratings Change')
plt.show()
```

```python
plt.plot([1,2,3],[y1,y2,y3])       # 1 stands for Year lies between 1922-1950  
plt.xlabel('Year')                 # 2 stands for Year lies between 1951-2000
plt.ylabel('Ratings')              # 3 stands for Year lies between 2001-2022 
plt.grid()
plt.title('Ratings Change')
plt.show()
```
